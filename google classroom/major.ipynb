## Cleaning the Data

Removing null values.

```python
new_df = df.dropna()
```

```python
new_df.info()
```

After removing all the nulls we see that there are 5975 rows left in the Dataframe.

```python
new_df.describe()
```

```python
new_df['Fuel_Type'].unique()
```

```python
new_df[new_df['Seats'] == 0]
```

```python
new_df['Seats'].unique()
```

```python
new_df['Owner_Type'].unique()
```

```python
new_df[new_df['Power'] == 'null bhp']
```

```python
new_df.info()
```

```python
new_df.describe()
```

```python
new_df = new_df[new_df['Power']!='null bhp']
```

```python
new_df.describe()
```

```python
new_df["Power"]= new_df["Power"].str.split(" ", n=1, expand = True) 
```

```python
new_df["Engine"]= new_df["Engine"].str.split(" ", n=1, expand = True) 
new_df["Mileage"]= new_df["Mileage"].str.split(" ", n=1, expand = True) 
```

```python
new_df.tail()
```

```python
new_df["Power"] = pd.to_numeric(new_df["Power"], downcast="float")
new_df["Engine"] = pd.to_numeric(new_df["Engine"], downcast="float")
new_df["Mileage"] = pd.to_numeric(new_df["Mileage"], downcast="float")
```

```python
new_df.tail()
```

```python
new_df.info()
```

```python
new_df.describe()
```

```python
new_df.to_excel('new_train.xlsx')
```
